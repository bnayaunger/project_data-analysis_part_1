```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.ad.co.il/nadlanrent?sp275=17413&sp277=18514,18129,17884,18183"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'} # This is chrome, you can set whatever browser you like

response = requests.get(url,headers= headers)
if response.status_code == 200:
    print("Success")
    result_page = BeautifulSoup(response.content,'html.parser')
else:
    print("Failure")
```

```python
prep_steps = result_page.find_all(lambda tag: tag.name == "div" and tag.get("class") == ["card-block"])
```

```python
rents=list()
for prep_step in prep_steps:
    rents.append("https://www.ad.co.il"+prep_step.find("a").get("href"))
```

```python
address = []
price = []
property_type=[]
neighborhood=[]
room_num=[]
floor_in_building = []
area = []
garden_area = []
days_to_enter = []
num_of_payment = []
monthly_arnona = []
building_tax = []
for rent in rents:
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
    response = requests.get(rent, headers=headers)
    result_page = BeautifulSoup(response.content, 'html.parser')
    div = result_page.find("div", {"class": "d-flex justify-content-between"})
    h2 = div.find_all('h2', {'class': 'card-title'})

    if len(h2) >= 2:  # בדיקה אם יש לפחות שני אלמנטים h2
        address.append(h2[0].get_text())
        price_text = h2[1].get_text().replace("₪", "").strip().replace(",","") # הסרת ₪ ורווחים
        try:
            price.append(float(price_text))
        except ValueError:
            print(f"לא ניתן להמיר מחיר ל-float: {price_text}")
            price.append(None) # או ערך ברירת מחדל אחר

    else:
        print(f"אין מספיק אלמנטים h2 עבור {rent}")
        address.append(None) # או ערך ברירת מחדל אחר
        price.append(None)   # או ערך ברירת מחדל אחר
    table = result_page.find("table", {'class': "table table-sm mb-4"})
    tbody = table.find("tbody")
    rows = tbody.find_all('tr')
    found_property_type = False
    found_neighborhood = False
    found_room_num = False
    found_floor = False
    found_area = False
    found_garden_area = False 
    found_days_to_enter = False
    found_num_of_payment = False
    found_monthly_arnona = False
    found_building_tax = False
    for row in rows:
        cells = row.find_all('td')
        if len(cells) == 2:  # וודא שיש שני תאים בכל שורה
            label = cells[0].text.strip()
            value = cells[1].text.strip()

            if label == 'פרטי הנכס':
                property_type.append(value)
                found_property_type = True
            elif label == 'שכונה':
                neighborhood.append(value)
                found_neighborhood = True
            elif label ==  'חדרים':
                room_num.append(float(value))
                found_room_num = True
            elif label == 'קומה':
                floor_in_building.append(value)
                found_floor = True 
            elif label ==  'שטח בנוי':
                area.append(value)
                found_area = True
            elif label ==  'שטח גינה':
                garden_area.append(value)
                found_garden_area = True
            elif label ==  'תאריך כניסה':
                days_to_enter.append(value)
                found_days_to_enter = True
            elif label ==  'תשלומים בשנה':
                num_of_payment.append(value)
                found_num_of_payment = True
            elif label ==  'ארנונה בחודש':
                monthly_arnona.append(value)
                found_monthly_arnona = True
            elif label ==  'ועד בית בחודש':
                building_tax.append(value)
                found_building_tax = True
    if not found_property_type:
        property_type.append(None)
    if not found_neighborhood:
        neighborhood.append(None)
    if not found_room_num:
        room_num.append(None)
    if not found_floor:
        floor.append(None)
    if not found_area:
        area.append(None)
    if not found_garden_area:
        garden_area.append(None)
    if not found_days_to_enter:
        days_to_enter.append(None)
    if not found_num_of_payment:
        num_of_payment.append(None)
    if not found_monthly_arnona:
        monthly_arnona.append(None)
    if not found_building_tax:
        building_tax.append(None)

```
